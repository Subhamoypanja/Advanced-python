```python
import matplotlib.pyplot as plt 
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df=pd.read_csv("Data\movie_score.csv")
df
```

```python
movie_scores=df.melt(id_vars='MovieTitle',value_vars=['Tomatometer','AudienceScore'],var_name='Type',value_name='Score')
```

```python
movie_scores
```

```python
#sns.set()
plt.figure(figsize=(10, 5), dpi=300)
# Create bar plot
ax = sns.barplot(x="MovieTitle", y="Score", hue="Type", data=movie_scores)
plt.xticks(rotation=10)
# Add title
plt.title("Movies Scores comparison")
plt.xlabel("Movies")
plt.ylabel("Scores")
# Show plot
plt.show()
```

```python
plt.style.use("classic")
plt.figure(figsize=(10,5),dpi=300)
y=movie_scores['Score'].values()
sns.barplot(x='MovieTitle',y='Score',hue='Type',data=movie_scores)
plt.xticks(rotation=10)
```
